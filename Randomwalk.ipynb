```python
prob = [0.5, 0.5]   
  

start = 0  
positions = [start] 
  
 
rr = np.random.random(100) 
downp = rr < prob[0] 
upp = rr > prob[1] 
  
  
for idownp, iupp in zip(downp, upp): 
    down = idownp and positions[-1] > -1
    up = iupp and positions[-1] < 1
    positions.append(positions[-1] - down + up) 
  

plt.plot(positions,'o') 
plt.show() 
```

```python
import pylab

n = 100000

x = np.zeros(n) 
y = np.zeros(n) 


for i in range(1, n): 
	val = random.randint(1, 4) 
	if val == 1: 
		x[i] = x[i - 1] + 0.5
		y[i] = y[i - 1] 
	elif val == 2: 
		x[i] = x[i - 1] - 0.5
		y[i] = y[i - 1] 
	elif val == 3: 
		x[i] = x[i - 1] 
		y[i] = y[i - 1] + 0.5
	else: 
		x[i] = x[i - 1] 
		y[i] = y[i - 1] - 0.5
	


pylab.plot(x, y) 
pylab.show() 
```

```python
import pylab

n = 1000


r = np.zeros(n) 
theta = np.zeros(n) 

 
for i in range(1, n): 
	val = random.randint(1, 2) 
	if val == 1: 
		r[i] = 10 * np.random.random()
		theta[i] = theta[i - 1] 
	elif val == 2: 
		r[i] = r[i - 1] 
		theta[i] = 10 * np.random.random() 



x = r*np.sin(theta)
y = r*np.cos(theta)

pylab.plot(x,y,'o') 
pylab.show() 
```
